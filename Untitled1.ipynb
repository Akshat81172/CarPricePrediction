## Data cleaning

```python
df['name']=df['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python

df['year'] = pd.to_numeric(df['year'], errors='coerce')

df= df.dropna(subset=['year'])
df['year'] = df['year'].astype(int)
print(df['year'].unique())
```

```python
df['year'] = df['year'].astype(int)
```

```python
df['Price'].unique()
```

```python
df['Price'] = df['Price'].replace('Ask For Price', np.nan)
df['Price'] = df['Price'].replace({',': ''}, regex=True)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df
```

```python
df['Price'].fillna(df['Price'].mode()[0],inplace=True)
```

```python
df['Price'] = df['Price'].astype(int)
```

```python
df.info()
```

```python
df["kms_driven"].unique()
```

```python
df['kms_driven'] = df['kms_driven'].replace('Petrol','20000')
```

###  Data train test split

```python
df
```

```python
x=df[['name','company','year','kms_driven','fuel_type']]
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
```

```python
oh=OneHotEncoder()
oh.fit(x[['name','company','fuel_type']])
```

```python
lr=LinearRegression()
```

```python
oh.categories_
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=oh.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
import pickle

# Save the trained model
with open('carpredict.pkl', 'wb') as model_file:
    pickle.dump(pipe, model_file)
```

```python
z=pipe.predict(x_test)
```

```python
z
```

```python
mean_absolute_error(y_test,z)
```

```python
mean_squared_error(y_test,z)
```

```python
r2_score(y_test,z)
```

```python
scores=[]
for i in range(100):
    x_train,y_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    z=pipe.predict(x_test)
    scores.append(r2_score(y_test,z))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
z=pipe.predict(x_test)
r2_score(y_test,z)
```

```python
input_values = [['Toyota Corolla Altis','Toyota',2022,20000,'Petrol']]
columns = pipe.feature_names_in_
input_data = pd.DataFrame(input_values, columns=columns)
q= pipe.predict(input_data)

```
